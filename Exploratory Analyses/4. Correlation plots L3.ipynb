```python
import pandas as pd
import matplotlib.pyplot as plt

vine_random_numbers = pd.read_parquet("../Data/vine_random_numbers_L3.parquet.gzip")
gaussian_random_numbers = pd.read_parquet("../Data/gaussian_random_numbers_L3.parquet.gzip")
t_random_numbers = pd.read_parquet("../Data/t_random_numbers_L3.parquet.gzip")
```

```python
cor_matrix = gaussian_random_numbers.sample(10000, random_state=100).corr()
plt.imshow(cor_matrix, cmap='viridis', interpolation='none', vmin=0, vmax=1)
plt.colorbar()
plt.title('Gaussian Copula', fontname="times new roman", fontsize=14, fontweight='bold')
plt.xlabel('Basin Index', fontname="times new roman", fontsize=12)
plt.ylabel('Basin Index', fontname="times new roman", fontsize=12)
plt.savefig('L3_gaus_copula.png', transparent=True)
```

```python
cor_matrix = t_random_numbers.corr()
plt.imshow(cor_matrix, cmap='viridis', interpolation='none', vmin=0, vmax=1)
plt.colorbar()
plt.title('T Copula', fontname="times new roman", fontsize=14, fontweight='bold')
plt.xlabel('Basin Index', fontname="times new roman", fontsize=12)
plt.ylabel('Basin Index', fontname="times new roman", fontsize=12)
plt.savefig('L3_t_copula.png', transparent=True)
```

```python
cor_matrix = vine_random_numbers.corr()
plt.imshow(cor_matrix, cmap='viridis', interpolation='none', vmin=0, vmax=1)
plt.colorbar()
plt.title('Vine Copula', fontname="times new roman", fontsize=14, fontweight='bold')
plt.xlabel('Basin Index', fontname="times new roman", fontsize=12)
plt.ylabel('Basin Index', fontname="times new roman", fontsize=12)
plt.savefig('L3_vine_copula.png', transparent=True)
```
